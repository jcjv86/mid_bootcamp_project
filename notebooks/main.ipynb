```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('../data/cleaned/airplanes_clean.csv')
data.head()
```

```python
aboard_deaths_op = data.groupby(['operator']).agg({'fatalities':'sum'}).rename(columns={'fatalities':'pax_deaths_by_operator'}).sort_values(by='pax_deaths_by_operator', ascending=False).reset_index()
aboard_deaths_op
```

```python
pax_deaths_type = data.groupby(['type']).agg({'fatalities':'sum'}).rename(columns={'fatalities':'pax_deaths_by_model'}).sort_values(by='pax_deaths_by_model', ascending=False).reset_index()
pax_deaths_type
```

```python
pax_deaths_type_n_operator = data.groupby(['type', 'operator']).agg({'fatalities':'sum'}).rename(columns={'fatalities':'total_deaths'}).sort_values(by='total_deaths', ascending=False).reset_index()
pax_deaths_type_n_operator
```

```python
pax_deaths_max = data.groupby(['type', 'operator', 'date', 'location']).agg({'fatalities':'sum'}).rename(columns={'fatalities':'total_passenger_deaths'}).sort_values(by='total_passenger_deaths', ascending=False).reset_index()
pax_deaths_max.head(10)
```

```python
ground_deaths = data.groupby(['type', 'operator', 'date', 'location', 'summary']).agg({'ground':'sum'}).rename(columns={'ground':'total_ground_casualties'}).sort_values(by='total_ground_casualties', ascending=False).reset_index()
ground_deaths
```
