```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('../data/cleaned/airplanes_clean.csv')
data.head()
```

```python
aboard_deaths_op = data.groupby(['operator']).agg({'fatalities':'sum'}).rename(columns={'fatalities':'deaths by operator'}).sort_values(by='deaths by operator', ascending=False).reset_index()
deadliest_operator_total = aboard_deaths_op.nlargest(50, ['deaths by operator'], keep='first')
deadliest_operator_total
```

```python
deadliest_operator_total.to_csv('../data/cleaned/deadliest_operator_total.csv', index = False)
```

```python
storm_caused = data[data['summary'].str.contains('storm')]
storm_caused['total_deaths'] = storm_caused['fatalities'] + storm_caused['ground']
storm_total_deaths = storm_caused['total_deaths'].sum()
display(storm_caused.head())
display(storm_caused.shape)
storm_caused_percent = round((storm_caused.shape[0]/data.shape[0])*100,2)
print('The percentage of accidents related to a storm is equal to',storm_caused_percent,'%')
print('The total number of casualties is',storm_total_deaths)
print('This represents the',round((storm_total_deaths/data_total_deaths)*100,2), '% of total deaths')
```

```python
sabotage_caused = data[data['summary'].str.contains('sabotage')]
sabotage_caused['total_deaths'] = sabotage_caused['fatalities'] + sabotage_caused['ground']
sabotage_total_deaths = sabotage_caused['total_deaths'].sum()
display(sabotage_caused.head(15))
display(sabotage_caused.shape)
sabotage_caused_percent = round((sabotage_caused.shape[0]/data.shape[0])*100,2)
print('The percentage of accidents related to a sabotage is equal to',sabotage_caused_percent,'%')
print('The total number of casualties is',sabotage_total_deaths)
print('This represents the',round((sabotage_total_deaths/data_total_deaths)*100,2), '% of total deaths')
```

```python
bomb_caused = data[data['summary'].str.contains('bomb')]
bomb_caused['total_deaths'] = bomb_caused['fatalities'] + bomb_caused['ground']
bomb_total_deaths = bomb_caused['total_deaths'].sum()
display(bomb_caused.head())
display(bomb_caused.shape)
bomb_caused_percent = round((bomb_caused.shape[0]/data.shape[0])*100,2)
print('The percentage of accidents related to a bomb is equal to',bomb_caused_percent,'%')
print('The total number of casualties is',bomb_total_deaths)
print('This represents the',round((bomb_total_deaths/data_total_deaths)*100,2), '% of total deaths')
```

```python
hijack_caused = data[data['summary'].str.contains('hijack')]
hijack_caused['total_deaths'] = hijack_caused['fatalities'] + hijack_caused['ground']
hijack_total_deaths = hijack_caused['total_deaths'].sum()
display(hijack_caused.tail())
display(hijack_caused.shape)
hijack_caused_percent = round((hijack_caused.shape[0]/data.shape[0])*100,2)
print('The percentage of accidents related to the flight being hijacked is equal to',hijack_caused_percent,'%')
print('The total number of casualties is',hijack_total_deaths)
print('This represents the',round((hijack_total_deaths/data_total_deaths)*100,2), '% of total deaths')
```

```python
non_sabotaged = data[~data['summary'].str.contains('hijack')]
non_sabotaged = non_sabotaged[~non_sabotaged['summary'].str.contains('bomb')]
non_sabotaged = non_sabotaged[~non_sabotaged['summary'].str.contains('sabotage')]
display(non_sabotaged.head())
display(non_sabotaged.shape)
non_sabotaged['total_deaths'] = non_sabotaged['fatalities'] + non_sabotaged['ground']
non_sabotaged_total_deaths = non_sabotaged['total_deaths'].sum()

non_sabotaged_percent = round((non_sabotaged.shape[0]/data.shape[0])*100,2)
print('The percentage of accidents non related to a bomb, sabotage or being hijacked is equal to',non_sabotaged_percent,'%')
print('The total number of casualties is',non_sabotaged_total_deaths)
print('This represents the',round((non_sabotaged_total_deaths/data_total_deaths)*100,2), '% of total deaths')
```
